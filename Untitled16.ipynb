```python
# Lab task 3 by dr Yousef Asiri for Computer Vision class (light and colors)
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the image
image = cv2.imread('/content/images.jpeg')
# Simulate different lighting conditions
low_light = cv2.convertScaleAbs(image, alpha=0.5, beta=0) # Dim lighting
normal_light = cv2.convertScaleAbs(image, alpha=1.0, beta=0) # Normal lighting
bright_light = cv2.convertScaleAbs(image, alpha=1.5, beta=0) # Bright lighting
# Display results
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(low_light, cv2.COLOR_BGR2RGB))
plt.title('Low Light')
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(normal_light, cv2.COLOR_BGR2RGB))
plt.title('Normal Light')
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(bright_light, cv2.COLOR_BGR2RGB))
plt.title('Bright Light')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load an image and convert to grayscale
image = cv2.imread('/content/images.jpeg', cv2.IMREAD_GRAYSCALE)
# Create a gradient to simulate shading
rows, cols = image.shape
gradient = np.linspace(0.3, 1, cols)
shading_effect = (image * gradient).astype(np.uint8)
# Display the original and shaded image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(shading_effect, cmap='gray')
plt.title('Shading Effect')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the image
image = cv2.imread('/content/images.jpeg')

# Create distortion coefficients
dist_coeffs = np.array([-0.3, 0.1, 0, 0], dtype=np.float64)
 # Adjust these fordifferent distortion effects
# Create a camera matrix (assume a standard focal length and center point)
h, w = image.shape[:2]
camera_matrix = np.array([[w, 0, w // 2], [0, h, h // 2], [0, 0, 1]], dtype=np.float32)
# Apply distortion
distorted_image = cv2.undistort(image, camera_matrix, dist_coeffs)
# Display the distorted image
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(distorted_image, cv2.COLOR_BGR2RGB))
plt.title('Distorted Image')
plt.show()
```
